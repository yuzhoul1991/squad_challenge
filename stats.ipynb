```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
context_file = os.path.join(os.getcwd(), 'data', 'train.context')
question_file = os.path.join(os.getcwd(), 'data', 'train.question')
```

```python
context_len = []
question_len = []
larger_than_len = 0
with open(context_file, 'r') as fh:
    line = fh.readline()
    while line:
        count = len(line.split())
        if count > 500: larger_than_len += 1
        context_len.append(count)
        line = fh.readline()
        
with open(question_file, 'r') as fh:
    line = fh.readline()
    while line:
        count = len(line.split())
        question_len.append(count)
        line = fh.readline()
```

```python
print(larger_than_len)
```

```python
plt.hist(context_len, 100)
```

```python
plt.hist(question_len, 50)
```

```python
span_file = os.path.join(os.getcwd(), 'data', 'train.span')
```

```python
start_idx = []
end_idx = []
with open(span_file, 'r') as fh:
    line = fh.readline()
    while line:
        start, end = line.split()
        start_idx.append(int(start))
        end_idx.append(int(end))
        line = fh.readline()
```

```python
print(start_idx)
```

```python
plt.hist(start_idx, 50)
```

```python
plt.hist(end_idx, 100)
```

```python
span_len = [end - start for start, end in zip(start_idx, end_idx)]
larger_len = sum(length > 30 for length in span_len)
```

```python
print(larger_len)
```

```python
plt.hist(span_len, 100)
```
